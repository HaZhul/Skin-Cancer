```python
%autoreload


import cv2
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import preprocess
import time
import keyboard
from IPython.display import clear_output
```

```python
def plot_before_and_after(image_before, image_after):
    plt.subplot(1, 2, 1)
    plt.title("Before")
    plt.imshow(image_before)
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.title("After")
    plt.imshow(image_after)
    plt.axis("off")
    
    plt.show()
```

```python
image_path = 'C:/Users/marci/Documents/SIIUM_sem_2/AOC/archive/ISIC_2019_Training_Input/ISIC_2019_Training_Input/ISIC_0025927.jpg'
preprocess_test_datasets = ['ISIC_0000000', 'ISIC_0000001', 'ISIC_0000002', 'ISIC_0000004', 'ISIC_0000072', 'ISIC_0000095', 'ISIC_0000074', 'ISIC_0025185', 'ISIC_0026616']


#PREPROCESSING

image_BGR = cv2.imread(image_path)
image_RGB = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2RGB)
image_G = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2GRAY)
image_HSV = cv2.cvtColor(image_BGR, cv2.COLOR_BGR2HSV)

mask = preprocess.run_preprocessing(image_G)
masked_image = image_HSV * mask[:, :, np.newaxis]

#PLOTTING

plot_before_and_after(image_RGB, masked_image)
```
