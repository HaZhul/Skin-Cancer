```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

from seg_mask import Segmentation_Mask

image_path = 'ISIC_2019_Training_Input/ISIC_2019_Training_Input/ISIC_0000004.jpg'
segmentation_creator = Segmentation_Mask()

segmented_image = segmentation_creator.make_knn_segmentation(image_path, k=3)

plt.subplot(1, 2, 1)

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Segmented Image")
plt.imshow(cv2.cvtColor(segmented_image, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.show()
```
